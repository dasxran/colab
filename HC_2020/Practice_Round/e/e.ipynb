```
line1 = "505000000 10000" #@param {type:"string"}
line2 = "223 225 245 249 252 276 307 341 343 350 378 380 389 422 428 463 485 502 515 536 558 567 584 611 614 615 641 644 672 713 737 747 766 776 786 842 849 873 874 902 918 944 991 1007 1009 1021 1051 1068 1111 1127 1130 1144 1170 1171 1201 1278 1282 1282 1296 1322 1326 1335 1338 1367 1399 1420 1424 1424 1436 1467 1519 1532 1552 1553 1559 1562 1567 1577 1587 1596 1606 1634 1654 1680 1708 1722 1724 1733 1745 1746 1747 1785 1801 1808 1865 1872 1874 1884 1904 1922 1924 1927 1930 1934 1936 1937 1964 1975 1991 2012 2065 2071 2086 2086 2093 2094 2095 2104 2112 2122 2127 2148 2152 2216 2227 2251 2263 2265 2272 2290 2305 2314 2319 2323 2329 2344 2351 2381 2385 2407 2411 2419 2422 2422 2438 2508 2523 2524 2533 2541 2595 2595 2597 2609 2630 2635 2640 2642 2655 2664 2664 2678 2690 2693 2711 2713 2716 2735 2736 2738 2749 2754 2786 2794 2808 2862 2893 2906 2929 2966 2968 2973 2977 2981 3004 3008 3015 3027 3096 3123 3139 3162 3176 3188 3198 3198 3210 3242 3242 3253 3291 3299 3307 3310 3318 3335 3336 3341 3364 3396 3401 3416 3422 3422 3435 3436 3441 3442 3457 3508 3517 3521 3543 3548 3552 3565 3582 3599 3631 3685 3688 3739 3753 3778 3782 3785 3801 3816 3838 3844 3849 3866 3891 3893 3901 3909 3924 3966 4003 4017 4061 4069 4096 4131 4136 4150 4200 4220 4224 4229 4298 4300 4302 4305 4318 4324 4337 4362 4365 4376 4405 4452 4458 4465 4469 4474 4504 4509 4542 4560 4562 4576 4585 4628 4694 4721 4756 4800 4827 4830 4833 4859 4897 4897 4936 4953 4980 5017 5021 5077 5082 5082 5084 5093 5107 5108 5113 5120 5130 5165 5174 5182 5186 5198 5202 5214 5282 5285 5306 5320 5321 5346 5368 5376 5377 5386 5394 5400 5419 5425 5457 5464 5504 5531 5543 5549 5578 5631 5636 5639 5655 5668 5704 5715 5718 5725 5745 5760 5761 5798 5799 5832 5834 5835 5854 5889 5907 5915 5920 5935 5945 5945 5952 5977 6002 6030 6085 6094 6111 6121 6149 6160 6164 6222 6237 6256 6261 6276 6293 6299 6340 6404 6412 6413 6427 6468 6488 6498 6507 6546 6600 6603 6639 6677 6680 6698 6701 6703 6731 6734 6737 6757 6761 6825 6850 6852 6873 6932 6932 6947 6959 6981 6993 6994 7006 7009 7033 7038 7068 7083 7085 7091 7112 7157 7186 7192 7204 7213 7230 7232 7236 7236 7252 7256 7270 7305 7323 7357 7370 7370 7396 7436 7463 7507 7512 7526 7535 7579 7590 7595 7608 7638 7657 7690 7724 7740 7743 7755 7768 7770 7778 7789 7789 7790 7801 7809 7820 7833 7840 7853 7861 7871 7882 7896 7925 7929 7929 7967 7977 7982 7992 8003 8012 8069 8092 8134 8140 8212 8237 8280 8290 8313 8325 8335 8348 8364 8366 8372 8375 8378 8399 8435 8477 8484 8486 8492 8506 8520 8520 8522 8532 8541 8581 8605 8614 8615 8644 8646 8649 8650 8669 8689 8690 8702 8737 8748 8757 8771 8776 8777 8797 8814 8826 8870 8875 8941 8947 8959 8972 8985 9002 9011 9014 9067 9082 9112 9124 9204 9230 9246 9284 9287 9294 9304 9323 9326 9339 9377 9416 9459 9461 9471 9488 9522 9559 9594 9596 9602 9611 9633 9645 9654 9727 9783 9796 9800 9810 9820 9827 9830 9838 9846 9868 9892 9896 9901 9904 9905 9911 9924 9955 9978 9995 9997 9999 10003 10022 10022 10059 10063 10090 10093 10096 10098 10103 10119 10127 10132 10138 10152 10229 10236 10241 10258 10296 10352 10367 10379 10402 10434 10446 10448 10461 10462 10474 10490 10511 10530 10535 10541 10552 10568 10583 10619 10621 10641 10650 10654 10659 10661 10671 10702 10707 10747 10753 10753 10754 10760 10773 10796 10801 10846 10853 10857 10869 10894 10915 10919 10953 10954 10959 10967 10981 10986 10988 10996 10999 11108 11151 11156 11165 11166 11174 11186 11203 11209 11228 11259 11261 11271 11272 11280 11290 11304 11304 11320 11346 11355 11388 11393 11422 11424 11503 11511 11546 11556 11568 11588 11591 11595 11599 11613 11642 11647 11654 11667 11676 11684 11685 11695 11696 11709 11710 11710 11719 11738 11743 11745 11753 11758 11777 11794 11797 11813 11823 11836 11873 11897 11901 11914 11914 11928 11965 11993 12022 12028 12061 12065 12073 12077 12088 12106 12114 12118 12124 12126 12131 12132 12132 12151 12158 12185 12185 12195 12201 12207 12212 12213 12230 12240 12244 12252 12254 12257 12296 12303 12310 12315 12374 12390 12406 12436 12491 12512 12516 12534 12552 12605 12661 12667 12707 12712 12713 12718 12719 12732 12773 12774 12822 12845 12891 12918 12933 12937 12954 12956 12958 12961 12962 12971 13024 13033 13054 13057 13068 13122 13123 13127 13211 13212 13216 13245 13251 13297 13329 13345 13365 13382 13400 13418 13427 13444 13449 13454 13459 13462 13466 13475 13498 13508 13517 13532 13545 13549 13558 13611 13651 13663 13669 13687 13706 13711 13717 13730 13732 13742 13776 13796 13800 13842 13846 13854 13855 13858 13860 13862 13866 13879 13885 13928 13937 13941 13952 13956 13961 13969 13980 13992 14008 14035 14055 14059 14061 14065 14111 14112 14116 14128 14152 14161 14162 14240 14254 14254 14272 14277 14304 14304 14311 14364 14373 14376 14389 14390 14395 14401 14410 14427 14428 14442 14445 14458 14470 14477 14489 14494 14497 14500 14505 14505 14505 14515 14533 14561 14563 14569 14626 14627 14635 14669 14704 14727 14734 14759 14785 14812 14826 14848 14857 14857 14860 14861 14864 14886 14891 14893 14895 14911 14912 14925 14926 14935 14967 14997 14998 15039 15084 15111 15112 15114 15186 15206 15212 15226 15237 15251 15262 15264 15273 15306 15307 15327 15336 15355 15380 15385 15408 15453 15484 15532 15536 15539 15560 15571 15574 15597 15602 15618 15634 15644 15662 15710 15737 15742 15761 15813 15834 15838 15870 15898 15902 15906 15909 15915 15924 15939 15943 15945 15958 15975 15990 15996 16047 16052 16058 16073 16090 16095 16109 16113 16114 16137 16146 16180 16180 16186 16187 16206 16283 16301 16302 16314 16318 16359 16367 16390 16420 16423 16454 16456 16462 16486 16504 16523 16527 16528 16555 16561 16601 16665 16675 16679 16702 16714 16742 16744 16761 16772 16774 16778 16786 16802 16804 16805 16812 16846 16849 16854 16862 16868 16890 16901 16915 16923 16935 16944 16960 16967 16970 16985 16999 17038 17047 17090 17101 17139 17153 17164 17207 17220 17224 17247 17253 17315 17330 17334 17336 17339 17360 17364 17371 17402 17404 17419 17436 17447 17479 17491 17506 17518 17532 17532 17538 17551 17558 17567 17598 17659 17680 17692 17697 17766 17767 17799 17810 17825 17835 17840 17851 17859 17869 17899 17904 17907 17909 17911 17912 17939 17942 17950 17972 17978 17983 18014 18016 18018 18073 18086 18088 18101 18115 18132 18134 18137 18148 18163 18186 18188 18188 18226 18234 18247 18256 18269 18277 18293 18330 18344 18393 18399 18407 18417 18434 18441 18442 18448 18448 18500 18501 18509 18534 18544 18545 18547 18577 18652 18659 18666 18697 18734 18740 18768 18775 18776 18780 18793 18811 18816 18860 18869 18874 18876 18903 18904 18914 18918 18920 18982 19008 19029 19059 19084 19088 19095 19106 19176 19214 19217 19229 19280 19292 19306 19314 19316 19330 19335 19345 19352 19375 19388 19402 19405 19466 19470 19476 19476 19523 19529 19534 19563 19590 19598 19613 19621 19642 19649 19652 19686 19690 19693 19715 19722 19779 19816 19817 19826 19830 19858 19858 19862 19872 19874 19878 19893 19925 19940 19945 19954 19962 19966 19997 20049 20094 20123 20128 20131 20223 20228 20245 20259 20263 20263 20297 20305 20315 20340 20342 20349 20356 20365 20369 20372 20404 20409 20429 20438 20512 20523 20523 20539 20578 20597 20605 20612 20618 20620 20631 20632 20667 20669 20675 20689 20691 20706 20709 20744 20755 20760 20773 20776 20791 20806 20817 20846 20857 20865 20873 20891 20898 20902 20915 20916 20918 20928 20939 20958 20965 20969 21012 21016 21034 21035 21117 21122 21161 21176 21182 21189 21189 21190 21197 21201 21217 21231 21238 21255 21263 21288 21292 21323 21334 21361 21379 21385 21423 21451 21473 21483 21493 21531 21541 21566 21587 21593 21636 21646 21662 21672 21680 21682 21689 21692 21706 21757 21761 21787 21806 21807 21808 21830 21839 21840 21857 21857 21863 21872 21876 21982 22019 22023 22024 22025 22035 22113 22115 22115 22173 22188 22197 22202 22207 22213 22228 22231 22286 22294 22302 22307 22325 22326 22374 22381 22390 22395 22404 22409 22416 22418 22420 22425 22429 22447 22457 22467 22471 22474 22478 22484 22503 22516 22523 22530 22544 22563 22578 22623 22643 22644 22654 22665 22668 22668 22676 22679 22696 22707 22712 22765 22769 22776 22795 22801 22801 22846 22846 22869 22877 22919 22925 22938 22954 22966 22991 22992 23003 23014 23027 23031 23068 23085 23088 23106 23144 23164 23172 23174 23183 23185 23191 23196 23214 23215 23222 23274 23275 23275 23280 23285 23303 23306 23386 23387 23408 23459 23498 23525 23525 23532 23547 23568 23568 23581 23607 23607 23610 23614 23618 23645 23652 23657 23677 23687 23692 23694 23696 23701 23719 23785 23806 23815 23815 23828 23830 23837 23841 23845 23852 23859 23859 23861 23871 23873 23874 23893 23896 23904 23909 23909 23933 23939 23968 23985 23998 24028 24032 24039 24040 24062 24079 24083 24091 24106 24108 24111 24115 24119 24141 24157 24161 24192 24232 24243 24254 24255 24293 24302 24303 24307 24312 24383 24396 24398 24439 24444 24462 24476 24513 24533 24558 24566 24567 24582 24587 24594 24597 24605 24607 24613 24617 24630 24635 24640 24643 24649 24669 24681 24692 24699 24711 24731 24746 24761 24762 24832 24917 24922 24927 24969 24975 24975 24987 25032 25034 25037 25104 25127 25152 25163 25168 25177 25199 25200 25205 25244 25246 25249 25255 25280 25283 25322 25323 25344 25355 25379 25396 25434 25445 25445 25449 25454 25498 25514 25524 25545 25551 25554 25557 25579 25587 25595 25599 25610 25625 25660 25671 25679 25697 25705 25712 25720 25724 25743 25787 25792 25800 25808 25812 25816 25819 25867 25877 25887 25890 25907 25944 25948 25957 25963 25998 26000 26007 26050 26050 26058 26070 26086 26088 26146 26155 26158 26170 26175 26180 26200 26214 26230 26240 26246 26268 26303 26311 26311 26339 26344 26365 26381 26406 26410 26427 26430 26450 26461 26467 26468 26476 26488 26510 26519 26534 26537 26548 26550 26553 26558 26611 26615 26627 26650 26652 26672 26675 26677 26685 26714 26727 26736 26745 26750 26753 26766 26789 26811 26866 26874 26890 26894 26905 26913 26915 26926 26994 27000 27011 27021 27038 27048 27049 27051 27065 27069 27088 27093 27094 27096 27108 27113 27120 27125 27147 27147 27150 27153 27176 27184 27204 27218 27228 27254 27260 27277 27278 27333 27347 27355 27359 27375 27388 27439 27444 27445 27450 27490 27492 27501 27546 27574 27607 27621 27654 27686 27703 27726 27771 27784 27785 27833 27849 27860 27867 27892 27898 27952 27958 27959 27969 27973 28006 28023 28036 28053 28057 28072 28074 28075 28077 28084 28095 28104 28110 28111 28132 28141 28141 28142 28159 28168 28189 28190 28201 28240 28243 28245 28254 28259 28264 28293 28329 28335 28353 28380 28416 28429 28451 28464 28470 28480 28484 28502 28514 28521 28522 28537 28538 28539 28542 28569 28603 28618 28625 28662 28679 28694 28707 28730 28745 28767 28788 28825 28834 28845 28846 28853 28857 28869 28870 28871 28874 28933 28944 28957 28966 29009 29017 29034 29046 29051 29071 29099 29119 29124 29130 29149 29152 29158 29211 29231 29253 29258 29261 29270 29278 29280 29284 29290 29290 29327 29347 29371 29385 29399 29437 29443 29443 29488 29526 29528 29538 29543 29567 29615 29621 29643 29658 29689 29697 29700 29730 29739 29756 29782 29782 29793 29804 29806 29813 29838 29866 29868 29878 29884 29912 29921 29940 29944 29969 29969 29993 30037 30041 30042 30056 30068 30073 30076 30077 30088 30091 30092 30108 30128 30132 30139 30146 30170 30187 30193 30206 30208 30225 30230 30248 30254 30264 30274 30288 30317 30352 30368 30369 30392 30393 30397 30435 30437 30453 30459 30480 30516 30525 30528 30536 30554 30566 30583 30589 30606 30608 30616 30626 30627 30637 30641 30658 30672 30680 30729 30757 30774 30786 30787 30801 30830 30831 30839 30850 30856 30858 30874 30954 30957 30962 30963 30973 30980 30984 30994 31004 31007 31017 31042 31064 31066 31071 31076 31089 31097 31098 31114 31148 31149 31149 31183 31188 31192 31214 31220 31235 31252 31263 31268 31288 31290 31299 31319 31334 31341 31344 31346 31375 31418 31423 31431 31453 31488 31489 31501 31512 31541 31549 31566 31610 31626 31628 31635 31652 31654 31658 31732 31732 31751 31754 31779 31799 31813 31816 31826 31830 31838 31851 31853 31853 31868 31869 31885 31911 31921 31934 31973 31980 31982 32014 32015 32032 32039 32048 32108 32127 32142 32149 32172 32180 32188 32196 32196 32219 32224 32232 32243 32255 32268 32319 32320 32349 32368 32382 32417 32438 32440 32456 32464 32470 32485 32487 32502 32506 32553 32566 32576 32587 32599 32600 32606 32616 32649 32652 32657 32665 32673 32721 32730 32746 32749 32753 32769 32799 32831 32832 32869 32870 32884 32885 32888 32894 32903 32907 32923 32923 32979 32979 32992 33013 33030 33054 33059 33112 33116 33118 33168 33185 33211 33226 33232 33263 33279 33281 33289 33291 33295 33304 33311 33327 33339 33351 33355 33383 33429 33446 33463 33465 33475 33525 33548 33568 33584 33584 33596 33602 33606 33607 33610 33633 33638 33643 33657 33671 33700 33701 33711 33719 33721 33725 33740 33807 33811 33819 33820 33843 33846 33855 33858 33884 33929 33942 33962 33969 33999 34000 34004 34026 34042 34053 34066 34101 34118 34123 34204 34219 34220 34255 34261 34277 34296 34297 34299 34304 34309 34331 34336 34360 34365 34378 34406 34408 34420 34435 34451 34465 34467 34480 34488 34494 34509 34515 34545 34554 34578 34583 34595 34602 34614 34617 34649 34651 34684 34691 34744 34750 34752 34781 34782 34789 34792 34799 34811 34845 34849 34850 34853 34859 34866 34881 34885 34886 34923 34936 34938 34945 34980 34985 34987 34990 34992 35016 35048 35058 35063 35104 35119 35149 35155 35163 35163 35173 35199 35200 35215 35226 35244 35256 35273 35297 35297 35311 35314 35351 35352 35380 35382 35398 35401 35404 35408 35414 35415 35429 35451 35456 35471 35487 35495 35518 35520 35551 35556 35612 35620 35620 35656 35670 35702 35718 35720 35744 35760 35764 35773 35781 35788 35788 35793 35829 35876 35881 35885 35931 35952 35958 35964 35965 35970 35978 35984 35985 35992 36016 36024 36056 36095 36101 36124 36148 36157 36170 36172 36176 36176 36178 36183 36186 36191 36201 36210 36237 36243 36249 36260 36261 36262 36271 36291 36296 36317 36341 36342 36375 36394 36398 36405 36424 36463 36474 36480 36481 36483 36483 36513 36528 36540 36584 36597 36603 36606 36617 36625 36630 36642 36657 36674 36685 36699 36700 36749 36780 36787 36804 36809 36842 36843 36846 36856 36863 36868 36882 36886 36888 36890 36893 36901 36904 36936 36948 36968 36978 36998 37018 37048 37061 37069 37099 37131 37225 37254 37257 37261 37275 37302 37309 37318 37334 37338 37340 37371 37392 37448 37452 37462 37470 37487 37488 37519 37521 37523 37530 37544 37546 37592 37601 37606 37612 37612 37629 37649 37653 37653 37710 37718 37729 37746 37751 37753 37755 37764 37783 37824 37838 37855 37862 37875 37876 37911 37916 37948 37958 37969 37990 37999 38019 38022 38040 38041 38063 38107 38108 38118 38130 38133 38165 38172 38188 38196 38230 38251 38251 38252 38297 38297 38310 38312 38348 38357 38361 38365 38375 38426 38433 38440 38453 38456 38459 38486 38497 38509 38526 38528 38537 38560 38586 38590 38598 38619 38664 38699 38700 38758 38827 38828 38832 38834 38843 38847 38853 38881 38885 38894 38913 38917 38922 38931 38939 38966 38967 38983 38984 38986 39001 39012 39022 39034 39041 39050 39065 39083 39085 39085 39087 39146 39147 39154 39157 39161 39163 39164 39169 39188 39222 39233 39244 39248 39252 39255 39257 39274 39298 39312 39317 39317 39317 39326 39334 39340 39344 39396 39404 39407 39412 39428 39483 39504 39512 39544 39546 39585 39620 39646 39653 39671 39701 39717 39726 39740 39748 39772 39773 39792 39818 39820 39831 39854 39882 39897 39943 39944 39960 39970 39975 40035 40074 40086 40147 40150 40154 40157 40176 40214 40225 40230 40239 40243 40256 40257 40266 40267 40274 40306 40313 40342 40357 40360 40388 40423 40436 40437 40439 40463 40485 40490 40493 40495 40506 40512 40549 40563 40577 40591 40605 40606 40608 40610 40622 40629 40630 40637 40641 40643 40648 40665 40705 40708 40716 40716 40758 40761 40762 40775 40780 40782 40818 40831 40839 40851 40865 40905 40932 40934 40940 40950 40959 40962 40975 40985 40998 41001 41021 41063 41093 41094 41109 41123 41176 41177 41180 41235 41302 41315 41338 41358 41366 41366 41372 41384 41412 41438 41451 41464 41465 41471 41474 41487 41489 41490 41518 41540 41553 41555 41573 41601 41604 41613 41615 41631 41632 41640 41661 41666 41718 41733 41778 41780 41815 41824 41874 41875 41887 41923 41932 41933 41940 41971 41975 41979 41985 42045 42054 42065 42069 42081 42117 42139 42140 42144 42151 42179 42203 42205 42225 42225 42234 42234 42241 42290 42297 42337 42351 42360 42381 42387 42389 42394 42440 42461 42470 42472 42493 42494 42502 42515 42518 42550 42558 42614 42616 42624 42659 42659 42668 42675 42683 42700 42733 42741 42763 42786 42797 42799 42799 42824 42839 42840 42889 42890 42906 42909 42914 42938 42968 43001 43011 43014 43020 43035 43065 43075 43112 43131 43136 43137 43148 43191 43203 43223 43230 43233 43234 43277 43285 43290 43318 43333 43335 43359 43364 43393 43412 43424 43450 43454 43456 43466 43484 43488 43560 43590 43592 43596 43613 43615 43670 43675 43696 43708 43712 43744 43749 43798 43805 43805 43806 43812 43812 43844 43881 43885 43921 43929 43943 43947 43955 43962 43969 43984 43991 43991 44014 44022 44033 44045 44053 44062 44073 44079 44082 44092 44112 44129 44136 44168 44179 44185 44210 44211 44238 44254 44255 44275 44282 44290 44295 44299 44302 44318 44340 44345 44360 44380 44399 44407 44408 44430 44439 44499 44524 44553 44570 44572 44629 44634 44656 44684 44707 44730 44748 44761 44761 44793 44797 44849 44878 44890 44892 44899 44899 44914 44919 44924 44931 44946 44961 44975 44983 44997 45056 45063 45080 45080 45096 45099 45121 45128 45138 45147 45185 45220 45220 45230 45244 45256 45342 45344 45355 45417 45428 45436 45461 45491 45501 45520 45526 45574 45599 45604 45613 45628 45659 45669 45677 45682 45692 45714 45722 45729 45738 45745 45760 45765 45774 45779 45821 45837 45852 45867 45878 45888 45932 45944 45955 45957 45974 45998 46001 46023 46027 46048 46079 46085 46093 46094 46099 46109 46111 46116 46126 46146 46147 46162 46168 46170 46180 46205 46214 46222 46241 46243 46245 46258 46259 46267 46278 46282 46308 46308 46334 46358 46364 46367 46374 46383 46423 46443 46452 46497 46500 46531 46554 46566 46610 46614 46620 46638 46649 46653 46662 46710 46711 46712 46721 46753 46767 46767 46786 46792 46813 46844 46846 46860 46865 46878 46927 46978 46986 46989 47004 47011 47030 47033 47051 47075 47089 47092 47098 47114 47120 47124 47130 47149 47162 47194 47202 47215 47217 47219 47222 47222 47233 47240 47269 47273 47276 47290 47319 47326 47333 47338 47357 47362 47365 47367 47370 47382 47387 47396 47401 47455 47480 47483 47490 47493 47495 47506 47520 47536 47544 47558 47561 47563 47569 47599 47608 47609 47674 47684 47698 47717 47742 47755 47769 47777 47783 47807 47840 47846 47869 47880 47882 47901 47907 47932 47958 47965 47984 47984 48017 48086 48102 48116 48139 48139 48185 48191 48193 48200 48210 48238 48265 48289 48301 48316 48367 48401 48403 48405 48412 48429 48433 48448 48452 48470 48499 48504 48530 48553 48565 48631 48636 48640 48640 48655 48655 48666 48668 48683 48704 48720 48730 48736 48837 48864 48891 48895 48905 48913 48939 48943 48961 48972 49001 49011 49041 49053 49061 49064 49069 49101 49112 49124 49141 49174 49185 49197 49211 49215 49247 49261 49261 49272 49304 49321 49324 49364 49374 49428 49443 49444 49466 49467 49467 49484 49503 49525 49526 49542 49543 49545 49546 49572 49576 49578 49593 49595 49610 49633 49653 49655 49659 49689 49699 49709 49725 49751 49762 49772 49839 49841 49849 49866 49867 49889 49894 49910 49914 49949 49958 49964 49969 49971 49974 49981 49986 50004 50014 50029 50042 50052 50056 50058 50063 50063 50115 50122 50134 50136 50151 50159 50160 50168 50176 50213 50218 50248 50271 50285 50297 50404 50412 50414 50425 50432 50449 50455 50490 50493 50502 50507 50510 50514 50534 50572 50579 50583 50585 50601 50608 50644 50654 50658 50666 50675 50678 50685 50697 50722 50732 50745 50775 50785 50792 50815 50819 50826 50828 50837 50862 50925 50925 50940 50955 50983 50996 51008 51031 51061 51061 51087 51097 51116 51142 51162 51162 51171 51174 51194 51199 51216 51228 51252 51299 51339 51343 51352 51358 51425 51429 51462 51490 51493 51506 51506 51560 51577 51597 51603 51607 51631 51660 51673 51694 51704 51750 51811 51852 51862 51914 51920 51928 51952 51983 51986 52040 52058 52059 52110 52133 52140 52144 52175 52191 52213 52227 52275 52279 52295 52302 52325 52334 52344 52356 52365 52367 52382 52390 52396 52399 52401 52414 52424 52435 52462 52476 52487 52487 52490 52494 52517 52536 52542 52544 52547 52550 52564 52577 52596 52604 52618 52626 52636 52649 52649 52676 52698 52707 52710 52711 52727 52775 52799 52805 52805 52812 52820 52838 52853 52862 52876 52886 52931 52941 52960 52980 52997 53013 53046 53068 53076 53077 53085 53090 53125 53132 53136 53136 53151 53153 53157 53171 53174 53179 53195 53209 53235 53237 53238 53245 53300 53322 53340 53355 53372 53377 53409 53416 53439 53447 53449 53454 53468 53504 53536 53562 53562 53572 53598 53608 53616 53626 53649 53657 53678 53684 53732 53736 53745 53746 53776 53779 53789 53812 53825 53865 53975 53981 53989 53991 53994 54023 54041 54045 54045 54050 54057 54070 54084 54092 54113 54117 54146 54151 54154 54174 54178 54224 54244 54244 54262 54328 54354 54361 54365 54394 54401 54405 54406 54421 54425 54430 54445 54449 54452 54496 54503 54506 54506 54517 54519 54537 54551 54554 54562 54586 54631 54631 54646 54670 54671 54755 54785 54789 54795 54844 54920 54960 54991 55002 55011 55049 55051 55089 55102 55122 55126 55142 55143 55159 55168 55175 55182 55189 55197 55209 55209 55210 55210 55218 55219 55244 55259 55265 55271 55299 55313 55352 55390 55401 55411 55414 55416 55435 55448 55456 55463 55474 55475 55508 55519 55531 55531 55563 55570 55574 55609 55615 55622 55648 55681 55686 55691 55706 55722 55725 55736 55742 55746 55769 55803 55806 55832 55864 55879 55925 55939 55943 55969 55973 55974 55976 56025 56052 56054 56073 56078 56083 56098 56130 56148 56221 56230 56233 56242 56280 56287 56296 56303 56320 56328 56350 56361 56371 56383 56419 56449 56487 56505 56508 56515 56567 56568 56596 56599 56617 56644 56684 56710 56711 56714 56716 56733 56744 56746 56756 56797 56811 56834 56860 56867 56867 56924 56940 56979 57000 57040 57053 57085 57090 57135 57150 57153 57156 57156 57176 57200 57207 57246 57269 57292 57297 57303 57315 57322 57339 57350 57350 57384 57403 57449 57456 57465 57490 57495 57496 57504 57515 57537 57560 57563 57568 57585 57596 57611 57615 57644 57682 57683 57694 57705 57716 57718 57719 57721 57748 57771 57794 57834 57837 57862 57897 57929 57960 57968 57971 57972 57988 58008 58011 58013 58021 58022 58027 58040 58046 58062 58079 58081 58095 58107 58143 58146 58164 58173 58186 58206 58226 58231 58271 58272 58278 58284 58291 58330 58336 58338 58367 58397 58406 58419 58421 58434 58434 58456 58463 58472 58490 58492 58503 58510 58513 58542 58577 58582 58632 58637 58639 58665 58707 58714 58721 58739 58745 58803 58860 58867 58870 58902 58917 58946 58979 58981 58984 58985 59001 59005 59048 59048 59056 59057 59062 59064 59089 59104 59104 59133 59142 59148 59166 59201 59202 59212 59273 59298 59306 59312 59327 59336 59336 59344 59348 59359 59363 59377 59384 59405 59408 59412 59415 59470 59487 59495 59509 59515 59517 59523 59587 59596 59626 59633 59650 59688 59741 59819 59834 59835 59842 59850 59852 59859 59863 59876 59902 59912 59943 59958 59960 60006 60017 60026 60075 60079 60099 60118 60155 60170 60212 60216 60218 60234 60244 60286 60325 60332 60359 60406 60416 60432 60434 60445 60457 60475 60491 60499 60510 60523 60524 60525 60540 60565 60584 60635 60709 60733 60733 60760 60798 60800 60812 60824 60868 60869 60880 60881 60932 60938 60945 60952 60980 60993 61045 61056 61066 61067 61076 61104 61125 61129 61152 61177 61181 61191 61195 61207 61218 61225 61230 61233 61237 61246 61276 61279 61279 61280 61290 61309 61376 61391 61408 61426 61427 61440 61453 61470 61474 61487 61496 61500 61543 61547 61580 61585 61610 61631 61643 61672 61675 61682 61705 61707 61731 61744 61747 61752 61763 61768 61770 61804 61858 61868 61883 61925 61932 61941 61972 61973 62008 62015 62050 62058 62065 62077 62092 62096 62101 62107 62118 62120 62137 62146 62148 62166 62167 62170 62195 62199 62211 62259 62260 62261 62263 62269 62303 62314 62355 62360 62365 62383 62430 62468 62472 62480 62485 62496 62497 62559 62566 62573 62577 62590 62592 62608 62611 62635 62648 62657 62685 62706 62717 62718 62732 62734 62741 62742 62757 62758 62771 62797 62802 62822 62858 62869 62897 62917 62919 62920 62927 62931 62940 62942 62971 63000 63009 63016 63030 63035 63080 63125 63136 63154 63194 63200 63206 63210 63213 63239 63249 63263 63290 63304 63306 63333 63339 63344 63349 63351 63369 63369 63399 63410 63417 63466 63471 63571 63582 63586 63593 63623 63623 63645 63654 63691 63705 63716 63722 63732 63751 63759 63792 63795 63804 63818 63834 63862 63864 63901 63914 63925 63941 63956 63973 63981 63984 63995 64045 64051 64070 64077 64083 64093 64114 64118 64128 64141 64160 64215 64223 64224 64238 64289 64308 64308 64351 64356 64360 64376 64380 64385 64394 64397 64406 64426 64459 64468 64486 64492 64501 64509 64544 64556 64580 64615 64617 64632 64669 64679 64729 64734 64744 64752 64771 64779 64806 64824 64843 64852 64859 64893 64895 64908 64942 64990 64993 64995 65013 65024 65043 65053 65054 65059 65072 65081 65084 65089 65095 65111 65116 65145 65146 65156 65168 65175 65187 65191 65225 65256 65256 65262 65270 65273 65275 65280 65281 65283 65296 65301 65301 65302 65353 65363 65384 65384 65402 65412 65428 65464 65466 65476 65504 65510 65513 65526 65597 65603 65613 65616 65633 65635 65652 65657 65658 65675 65686 65698 65699 65709 65732 65737 65743 65757 65758 65762 65772 65774 65833 65839 65844 65884 65893 65895 65895 65901 65908 65917 65917 65919 65970 65992 66030 66068 66088 66099 66104 66107 66152 66168 66182 66210 66234 66237 66245 66260 66260 66274 66330 66332 66363 66365 66370 66386 66399 66439 66454 66455 66462 66489 66502 66518 66526 66552 66605 66605 66611 66623 66689 66693 66695 66727 66779 66808 66832 66861 66861 66861 66865 66895 66905 66921 66924 66974 67011 67018 67039 67046 67051 67078 67090 67115 67133 67137 67167 67183 67187 67193 67254 67281 67289 67293 67301 67306 67369 67373 67373 67384 67426 67426 67464 67482 67509 67560 67564 67618 67641 67648 67657 67660 67662 67736 67750 67753 67759 67783 67786 67787 67804 67810 67816 67847 67896 67902 67926 67938 68004 68005 68006 68007 68015 68027 68049 68052 68095 68112 68120 68146 68161 68179 68188 68208 68237 68259 68269 68279 68294 68297 68307 68360 68399 68399 68402 68417 68434 68434 68464 68465 68474 68505 68520 68535 68536 68559 68561 68564 68570 68612 68618 68648 68654 68656 68659 68676 68686 68699 68707 68715 68765 68767 68781 68794 68824 68825 68827 68830 68871 68882 68925 68931 68975 69001 69028 69046 69060 69077 69096 69105 69131 69139 69143 69168 69201 69206 69222 69227 69236 69236 69246 69247 69249 69283 69326 69343 69402 69417 69439 69442 69469 69500 69501 69510 69527 69529 69530 69543 69543 69552 69576 69578 69581 69603 69637 69666 69670 69671 69675 69684 69688 69709 69721 69722 69724 69738 69765 69775 69791 69808 69830 69832 69834 69878 69882 69888 69935 69952 69984 69998 70045 70063 70069 70093 70120 70142 70147 70157 70176 70187 70187 70191 70208 70216 70238 70278 70283 70287 70290 70310 70313 70317 70317 70322 70365 70387 70415 70426 70438 70448 70536 70543 70559 70562 70576 70581 70590 70596 70602 70608 70643 70655 70663 70671 70696 70723 70737 70739 70759 70768 70782 70783 70793 70801 70822 70824 70827 70844 70851 70899 70918 70924 70942 70966 70966 70973 71001 71012 71065 71093 71094 71220 71228 71255 71256 71266 71276 71307 71309 71315 71323 71362 71371 71372 71404 71429 71460 71467 71468 71492 71499 71531 71539 71568 71595 71612 71642 71658 71675 71679 71690 71711 71711 71723 71749 71759 71769 71771 71810 71823 71832 71832 71837 71845 71852 71852 71861 71877 71884 71886 71896 71904 71922 71933 71935 71956 71969 71975 71992 72000 72022 72033 72037 72038 72067 72110 72114 72138 72146 72150 72154 72171 72174 72185 72205 72220 72222 72252 72280 72285 72294 72298 72330 72347 72347 72375 72389 72399 72406 72437 72460 72506 72520 72521 72536 72537 72538 72541 72542 72578 72605 72617 72623 72653 72664 72664 72679 72683 72685 72723 72730 72731 72752 72784 72806 72852 72871 72886 72890 72896 72899 72905 72913 72913 72918 72937 73007 73013 73017 73037 73043 73057 73059 73070 73106 73110 73118 73118 73127 73136 73144 73167 73196 73202 73212 73219 73225 73235 73257 73310 73320 73340 73368 73371 73376 73383 73449 73459 73461 73504 73511 73520 73528 73533 73542 73543 73555 73563 73620 73643 73652 73671 73680 73692 73695 73699 73721 73747 73759 73762 73799 73813 73845 73856 73863 73885 73898 73899 73945 73956 73972 73999 74000 74027 74061 74065 74072 74085 74088 74124 74137 74155 74180 74184 74193 74206 74221 74245 74303 74310 74313 74342 74345 74359 74368 74375 74379 74387 74398 74403 74421 74446 74452 74457 74457 74465 74490 74506 74514 74543 74548 74579 74593 74602 74630 74653 74662 74676 74680 74688 74695 74723 74738 74749 74750 74764 74776 74779 74790 74797 74809 74820 74827 74885 74891 74893 74895 74908 74928 74928 74930 74947 74976 74977 74995 75004 75004 75020 75063 75102 75108 75135 75151 75155 75175 75198 75203 75225 75297 75302 75311 75322 75335 75354 75372 75374 75388 75389 75404 75413 75448 75456 75460 75485 75500 75512 75529 75535 75601 75601 75601 75603 75615 75616 75643 75675 75688 75701 75706 75725 75725 75736 75755 75769 75787 75796 75817 75821 75856 75858 75864 75874 75903 75906 75917 75961 76015 76026 76029 76047 76062 76077 76082 76082 76083 76095 76115 76116 76132 76136 76165 76177 76179 76185 76211 76239 76273 76278 76317 76341 76350 76371 76373 76374 76377 76393 76407 76425 76439 76478 76484 76501 76506 76512 76531 76552 76556 76571 76578 76602 76614 76616 76620 76640 76658 76662 76682 76697 76705 76707 76710 76718 76742 76770 76779 76782 76816 76876 76913 76965 76988 77042 77053 77055 77105 77112 77113 77125 77140 77152 77156 77159 77210 77214 77235 77266 77284 77309 77316 77317 77319 77319 77333 77337 77381 77406 77407 77415 77420 77433 77458 77489 77506 77506 77509 77511 77527 77539 77543 77546 77561 77562 77570 77570 77575 77583 77593 77625 77632 77646 77659 77675 77679 77691 77701 77705 77709 77723 77740 77741 77762 77801 77860 77890 77909 77950 77956 77981 77989 78011 78141 78150 78162 78168 78171 78174 78175 78218 78227 78243 78309 78314 78326 78334 78372 78411 78413 78415 78415 78422 78435 78436 78444 78450 78463 78465 78471 78480 78527 78533 78540 78543 78548 78559 78608 78622 78628 78634 78643 78648 78654 78656 78659 78672 78681 78704 78706 78718 78742 78747 78749 78767 78774 78779 78789 78814 78820 78832 78837 78843 78853 78888 78895 78904 78913 78916 78918 78923 78928 78931 78947 78947 78951 78958 78970 78975 79001 79011 79037 79077 79084 79093 79125 79136 79140 79145 79146 79158 79197 79198 79202 79216 79229 79231 79245 79246 79253 79259 79271 79275 79276 79306 79306 79326 79328 79333 79354 79366 79393 79399 79400 79406 79418 79426 79438 79451 79454 79501 79504 79533 79540 79549 79556 79565 79578 79583 79584 79585 79586 79642 79645 79657 79677 79679 79691 79698 79721 79738 79780 79801 79849 79859 79867 79869 79870 79877 79911 79912 79914 79916 79925 79949 79968 79977 79986 79988 80036 80038 80049 80053 80061 80065 80072 80073 80098 80107 80107 80112 80131 80135 80135 80173 80185 80197 80204 80204 80254 80254 80275 80278 80288 80298 80304 80328 80331 80377 80416 80447 80471 80533 80546 80547 80561 80580 80588 80625 80628 80632 80701 80767 80780 80791 80813 80813 80839 80856 80870 80881 80891 80906 80935 80950 80957 80962 80963 80967 80985 80986 81009 81011 81015 81022 81026 81033 81051 81052 81083 81118 81163 81163 81168 81175 81185 81199 81203 81217 81273 81278 81279 81279 81300 81326 81339 81356 81376 81387 81401 81441 81485 81492 81507 81510 81526 81537 81547 81559 81568 81572 81575 81607 81648 81662 81675 81676 81714 81722 81722 81737 81760 81777 81826 81849 81878 81878 81894 81902 81915 81961 81978 81991 82014 82016 82018 82027 82056 82090 82105 82107 82110 82115 82120 82127 82133 82142 82161 82175 82177 82183 82225 82231 82235 82254 82267 82271 82300 82308 82322 82322 82325 82344 82344 82349 82358 82359 82360 82378 82410 82429 82433 82434 82441 82449 82480 82489 82490 82492 82500 82522 82562 82562 82576 82587 82591 82615 82641 82645 82646 82649 82660 82666 82691 82725 82751 82778 82816 82835 82885 82895 82895 82908 82918 82920 82932 82947 82956 82967 82967 82999 83012 83019 83054 83074 83084 83093 83101 83101 83125 83132 83135 83139 83144 83162 83166 83166 83173 83195 83202 83211 83216 83232 83268 83285 83306 83307 83311 83348 83374 83378 83383 83400 83430 83441 83457 83499 83506 83525 83550 83554 83561 83563 83577 83596 83609 83619 83642 83650 83689 83701 83734 83742 83743 83750 83755 83792 83799 83801 83855 83861 83864 83907 83907 83929 83960 83978 83979 83993 84003 84010 84021 84045 84070 84097 84123 84144 84148 84169 84200 84208 84213 84253 84255 84261 84261 84269 84280 84314 84317 84332 84364 84393 84404 84487 84503 84507 84510 84516 84516 84516 84553 84559 84565 84570 84617 84636 84640 84659 84667 84676 84691 84695 84711 84721 84768 84769 84770 84776 84784 84789 84826 84838 84873 84890 84918 84928 84937 84945 84969 84979 84981 84983 84991 84994 85014 85027 85031 85044 85051 85075 85115 85154 85155 85176 85192 85197 85204 85209 85228 85233 85236 85248 85270 85282 85302 85307 85317 85335 85337 85368 85370 85375 85401 85410 85421 85433 85446 85450 85454 85457 85474 85496 85512 85535 85577 85625 85630 85652 85655 85663 85680 85690 85693 85694 85707 85725 85740 85756 85758 85791 85848 85856 85861 85862 85877 85881 85910 85951 85959 85997 85999 86000 86010 86014 86029 86036 86074 86100 86130 86144 86163 86165 86175 86185 86214 86215 86216 86235 86252 86274 86304 86309 86328 86330 86332 86346 86356 86357 86365 86404 86417 86501 86505 86509 86521 86521 86540 86558 86588 86610 86615 86627 86657 86664 86673 86760 86767 86773 86774 86775 86777 86785 86794 86820 86843 86856 86866 86887 86915 86931 86939 86943 86967 86988 86995 86999 87008 87009 87014 87026 87026 87033 87041 87041 87044 87053 87100 87105 87105 87114 87118 87127 87158 87159 87161 87163 87168 87200 87202 87213 87225 87231 87245 87281 87289 87319 87348 87368 87397 87404 87422 87442 87458 87470 87485 87505 87512 87532 87556 87569 87570 87575 87589 87610 87628 87630 87640 87644 87649 87693 87702 87717 87831 87834 87836 87859 87886 87916 87927 87929 87942 87950 88000 88010 88022 88023 88030 88046 88047 88068 88069 88095 88128 88151 88185 88192 88201 88215 88241 88249 88252 88272 88315 88325 88328 88355 88379 88379 88385 88425 88428 88437 88481 88501 88551 88572 88574 88633 88645 88650 88651 88674 88677 88691 88699 88705 88731 88736 88741 88759 88762 88770 88788 88789 88792 88794 88828 88884 88885 88901 88906 88907 88916 88920 88933 88949 88952 88971 88985 88993 88999 89049 89062 89067 89077 89090 89107 89107 89114 89142 89147 89202 89209 89229 89241 89282 89283 89283 89297 89318 89321 89346 89360 89375 89380 89392 89395 89412 89414 89418 89485 89502 89515 89515 89517 89528 89540 89553 89564 89652 89659 89676 89679 89685 89686 89736 89757 89784 89787 89798 89803 89831 89852 89866 89891 89894 89906 89939 90020 90029 90030 90043 90045 90046 90061 90076 90079 90085 90089 90090 90133 90136 90136 90141 90160 90175 90187 90231 90259 90266 90267 90277 90280 90288 90310 90314 90323 90330 90338 90338 90339 90344 90348 90366 90387 90410 90415 90419 90443 90447 90470 90471 90481 90487 90513 90516 90517 90588 90619 90632 90649 90656 90660 90673 90675 90678 90685 90686 90700 90706 90717 90737 90788 90820 90837 90865 90904 90908 90944 90998 91013 91068 91086 91087 91103 91107 91107 91115 91209 91218 91240 91247 91248 91253 91276 91286 91289 91339 91366 91384 91388 91407 91408 91416 91452 91482 91505 91507 91507 91513 91514 91578 91578 91588 91597 91613 91637 91656 91663 91691 91694 91787 91794 91842 91847 91859 91861 91877 91878 91888 91904 91941 91981 92034 92039 92077 92087 92116 92165 92219 92219 92237 92239 92251 92262 92274 92284 92306 92323 92329 92348 92353 92353 92356 92364 92378 92382 92392 92412 92426 92438 92452 92460 92471 92506 92521 92528 92538 92554 92559 92560 92622 92662 92675 92689 92710 92711 92715 92736 92763 92769 92771 92814 92878 92893 92894 92915 92931 92954 92961 92971 92977 92985 93000 93043 93075 93077 93079 93115 93124 93139 93173 93210 93215 93216 93264 93290 93298 93325 93333 93336 93342 93361 93364 93382 93388 93397 93415 93425 93434 93456 93457 93474 93520 93531 93531 93549 93660 93675 93740 93744 93779 93797 93811 93822 93882 93887 93889 93899 93924 93969 93981 93998 94001 94018 94028 94038 94039 94070 94071 94092 94106 94112 94123 94154 94170 94178 94203 94205 94225 94227 94242 94253 94254 94257 94289 94299 94305 94320 94322 94327 94335 94340 94372 94374 94376 94378 94379 94379 94438 94449 94455 94463 94479 94504 94549 94551 94575 94605 94622 94626 94629 94630 94634 94639 94648 94653 94671 94684 94686 94721 94728 94731 94741 94756 94763 94781 94790 94803 94807 94815 94825 94851 94860 94867 94877 94890 94900 94900 94905 94935 94943 94958 94964 94992 95001 95002 95034 95037 95039 95040 95052 95058 95071 95096 95107 95136 95146 95155 95156 95167 95234 95234 95246 95253 95276 95311 95311 95321 95344 95354 95358 95359 95402 95436 95440 95446 95449 95473 95477 95505 95510 95511 95545 95552 95586 95594 95615 95618 95641 95672 95679 95693 95699 95750 95751 95792 95839 95847 95854 95887 95889 95904 95953 95961 95976 95991 96033 96038 96042 96052 96086 96087 96088 96091 96098 96108 96125 96129 96148 96231 96238 96253 96259 96267 96300 96371 96394 96399 96421 96429 96437 96492 96495 96498 96523 96527 96542 96550 96554 96555 96565 96588 96601 96601 96602 96628 96649 96675 96708 96717 96731 96741 96762 96763 96782 96819 96821 96823 96837 96840 96861 96863 96887 96904 96908 96910 96926 96937 96950 96950 96961 96988 96990 97009 97019 97037 97046 97070 97085 97097 97099 97100 97108 97133 97175 97208 97233 97241 97250 97293 97304 97320 97333 97335 97343 97379 97400 97413 97430 97447 97459 97461 97476 97507 97512 97560 97560 97578 97580 97609 97611 97635 97645 97677 97684 97692 97722 97755 97781 97799 97810 97832 97834 97840 97847 97881 97888 97930 97931 97936 97941 97958 97975 97984 97994 98007 98014 98018 98032 98033 98033 98055 98059 98060 98060 98090 98107 98127 98135 98144 98149 98195 98211 98214 98219 98220 98232 98232 98270 98274 98299 98303 98309 98310 98312 98321 98323 98363 98363 98380 98411 98413 98426 98427 98463 98482 98494 98512 98521 98543 98550 98555 98568 98585 98596 98602 98616 98625 98639 98644 98674 98679 98701 98707 98744 98756 98781 98801 98818 98839 98853 98855 98865 98887 98891 98912 98926 98940 98945 98952 98955 98983 99026 99031 99050 99051 99070 99073 99073 99081 99084 99109 99113 99132 99132 99159 99203 99217 99220 99222 99230 99238 99258 99342 99346 99385 99392 99405 99427 99431 99457 99458 99469 99478 99501 99530 99533 99537 99538 99546 99550 99564 99578 99580 99586 99591 99622 99626 99665 99672 99714 99735 99750 99763 99773 99786 99796 99815 99818 99834 99848 99855 99865 99866 99870 99882 99908 99913 99914 99915 99933 99966 99974 99989 99999 100007 100043 100047 100052 100061 100084 100096 100118 100118 100138 100139 100161 100173 100175 100180 100186 100187 100194 100197 100205 100221 100225 100229 100230 100240 100248 100253 100279 100282 100306 100335 100341 100342 100359 100366 100394 100437 100437 100465 100467 100470 100482 100482 100505 100516 100529 100531 100546 100547 100549 100563 100590 100594 100596 100606 100611 100613 100620 100645 100685 100730 100732 100746 100752 100755 100759 100767 100788 100816 100823 100843 100849 100853 100858 100861 100872 100874 100877 100887 100903 100903 100907 100949 100955 100975 100990 101020 101038 101038 101045 101046 101051 101065 101069 101073 101080 101089 101094 101094 101139 101177 101188 101213 101213 101247 101314 101356 101358 101360 101372 101409 101418 101432 101480 101502 101525 101536 101536 101557 101585 101608 101619 101647 101678 101695 101712 101737 101755 101757 101759 101771 101779 101788 101796 101804 101860 101921 101936 101937 101938 101943 101951 101984 102012 102022 102023 102034 102036 102044 102061 102067 102069 102080 102081 102082 102114 102131 102135 102137 102145 102173 102196 102202 102207 102216 102223 102300 102306 102322 102323 102338 102364 102396 102421 102424 102432 102433 102435 102455 102464 102501 102516 102534 102547 102556 102563 102599 102603 102616 102628 102659 102669 102697 102715 102722 102726 102734 102739 102785 102794 102800 102856 102891 102896 102911 102924 102939 102950 102981 103036 103072 103092 103104 103127 103145 103168 103183 103192 103196 103197 103206 103211 103211 103212 103217 103238 103241 103287 103293 103328 103353 103368 103386 103411 103426 103432 103460 103491 103507 103525 103530 103537 103556 103569 103570 103577 103596 103625 103628 103636 103656 103668 103682 103690 103696 103698 103730 103747 103747 103764 103777 103785 103816 103827 103829 103831 103841 103851 103867 103869 103900 103904 103908 103913 103929 103948 103953 103955 103959 103959 103983 103996 104022 104032 104074 104099 104106 104129 104175 104179 104181 104186 104194 104215 104244 104245 104257 104270 104275 104284 104290 104295 104312 104365 104380 104391 104394 104398 104401 104404 104417 104418 104446 104459 104500 104502 104526 104527 104543 104544 104568 104570 104576 104593 104597 104598 104611 104639 104656 104657 104672 104722 104725 104729 104759 104774 104778 104794 104846 104858 104886 104901 104910 104958 104967 104974 104974 105005 105026 105064 105074 105101 105120 105124 105140 105176 105177 105205 105211 105211 105213 105219 105222 105227 105232 105244 105262 105264 105269 105272 105277 105281 105301 105306 105329 105331 105363 105394 105422 105466 105482 105485 105491 105499 105509 105529 105532 105583 105585 105590 105600 105619 105652 105655 105661 105686 105687 105715 105716 105745 105753 105780 105805 105838 105911 105917 105941 105945 105947 105948 105954 105960 106048 106057 106084 106090 106095 106105 106132 106153 106158 106182 106190 106209 106212 106218 106223 106224 106239 106252 106254 106264 106309 106334 106377 106381 106393 106404 106442 106455 106463 106478 106487 106488 106492 106497 106502 106549 106568 106577 106590 106596 106638 106640 106651 106655 106686 106707 106730 106740 106743 106765 106766 106793 106827 106836 106861 106869 106895 106909 106913 106928 106938 106966 106967 106971 107023 107045 107050 107059 107153 107167 107186 107223 107233 107245 107247 107254 107258 107291 107293 107310 107318 107325 107357 107362 107363 107377 107383 107405 107453 107455 107468 107473 107491 107498 107531 107538 107550 107582 107616 107628 107674 107689 107705 107719 107748 107762 107778 107868 107881 107883 107888 107892 107899 107905 107905 107913 107915 107967 107968 107970 108007 108008 108015 108016 108030 108098 108101 108111 108123 108124 108128 108129 108135 108141 108185 108202 108202 108212 108217 108218 108257 108281 108293 108294 108307 108315 108319 108360 108375 108381 108399 108420 108441 108452 108456 108456 108471 108478 108481 108494 108499 108509 108526 108531 108545 108573 108582 108613 108613 108619 108665 108678 108679 108694 108710 108716 108725 108725 108733 108739 108739 108776 108779 108787 108810 108844 108844 108851 108858 108863 108889 108894 108956 108960 108962 108963 108974 108975 108993 109004 109031 109039 109065 109102 109104 109119 109122 109150 109153 109156 109169 109170 109187 109193 109205 109212 109215 109234 109240 109246 109248 109250 109260 109298 109311 109318 109320 109338 109362 109386 109389 109393 109394 109401 109401 109406 109455 109456 109473 109478 109494 109503 109551 109556 109576 109602 109605 109640 109665 109666 109695 109696 109717 109735 109747 109796 109812 109832 109835 109839 109841 109856 109863 109885 109925 109942 109950 109951 109959 109981 109983 109993 110040 110044 110056 110063 110072 110077 110098 110100 110102 110119 110136 110174 110175 110191 110195 110210 110219 110247 110253 110255 110257 110316 110317 110335 110339 110365 110396 110418 110425 110480 110527 110529 110529 110553 110565 110567 110578 110588 110590 110608 110700 110711 110742 110752 110757 110762 110777 110794 110794 110805 110826 110829 110830 110841 110867 110888 110908 110921 110927 110940 110946 110949 110966 110986 111008 111015 111021 111053 111069 111071 111073 111089 111098 111101 111106 111121 111124 111134 111154 111164 111186 111194 111198 111229 111235 111238 111238 111239 111270 111271 111285 111292 111299 111302 111306 111335 111345 111353 111354 111360 111374 111382 111383 111404 111431 111481 111534 111552 111561 111582 111589 111606 111618 111648 111650 111652 111702 111740 111744 111793 111795 111796 111800 111848 111873 111878 111893 111894 111901 111904 111929 111937 111946 111953 111971 111977 111996 111996 112048 112068 112074 112102 112132 112150 112155 112156 112164 112165 112166 112189 112221 112240 112282 112295 112304 112316 112320 112328 112337 112346 112367 112376 112388 112426 112427 112437 112442 112479 112488 112497 112512 112529 112531 112559 112606 112614 112616 112635 112648 112649 112667 112690 112702 112721 112723 112735 112745 112762 112768 112779 112802 112805 112829 112835 112867 112893 112900 112939 112940 112941 112945 112975 113029 113037 113042 113073 113077 113082 113084 113085 113094 113101 113102 113133 113160 113166 113171 113179 113198 113223 113249 113251 113263 113265 113269 113315 113372 113391 113391 113405 113407 113414 113430 113433 113437 113454 113467 113482 113499 113504 113541 113579 113590 113609 113659 113669 113685 113690 113715 113750 113822 113846 113873 113881 113883 113895 113897 113899 113918 113924 113946 113947 113961 113964 113990 113999 114013 114024 114036 114051 114057 114098 114100 114119 114147 114148 114162 114163 114172 114205 114218 114227 114229 114275 114283 114295 114326 114332 114335 114336 114361 114384 114410 114436 114440 114452 114455 114460 114465 114493 114510 114510 114513 114528 114528 114530 114553 114556 114573 114599 114608 114682 114688 114723 114745 114748 114762 114781 114785 114851 114881 114885 114889 114934 114935 114945 114947 114953 114956 114956 114974 115015 115035 115046 115049 115050 115062 115063 115065 115068 115077 115129 115139 115158 115170 115193 115198 115208 115241 115271 115272 115274 115300 115310 115322 115356 115375 115392 115406 115427 115440 115456 115475 115476 115482 115510 115517 115541 115565 115570 115592 115602 115624 115651 115665 115667 115695 115700 115716 115724 115735 115760 115777 115816 115836 115881 115902 115905 115942 115942 115958 115986 115994 115997 116012 116025 116026 116032 116035 116038 116047 116052 116079 116080 116083 116096 116106 116110 116117 116123 116123 116128 116163 116167 116169 116174 116200 116214 116218 116220 116257 116270 116320 116331 116333 116347 116356 116359 116360 116365 116372 116431 116436 116443 116478 116500 116508 116521 116533 116554 116556 116568 116577 116581 116589 116642 116645 116650 116653 116673 116675 116678 116725 116744 116756 116775 116785 116808 116814 116828 116834 116872 116884 116886 116901 116911 116922 116924 116948 116971 116975 116981 117001 117006 117016 117029 117046 117056 117058 117059 117074 117086 117094 117095 117102 117110 117112 117120 117131 117144 117160 117166 117175 117184 117188 117192 117232 117236 117260 117273 117282 117288 117310 117313 117361 117397 117413 117421 117440 117510 117667 117687 117712 117713 117726 117753 117767 117782 117799 117802 117812 117834 117855 117856 117862 117867 117878 117880 117884 117914 117921 117923 117948 117956 117958 117984 117987 118004 118008 118038 118052 118054 118059 118095 118095 118109 118127 118145 118147 118152 118158 118159 118177 118181 118184 118196 118241 118258 118275 118297 118298 118300 118302 118310 118335 118339 118356 118376 118398 118411 118421 118428 118460 118461 118467 118503 118555 118556 118568 118573 118587 118610 118622 118665 118667 118668 118668 118684 118687 118691 118701 118704 118711 118742 118788 118795 118797 118805 118807 118829 118845 118878 118888 118891 118911 118945 118978 118990 118990 119012 119020 119026 119049 119059 119060 119065 119068 119075 119083 119094 119110 119120 119121 119130 119137 119139 119141 119145 119147 119195 119222 119223 119228 119234 119235 119293 119295 119305 119317 119334 119380 119383 119388 119393 119396 119397 119402 119413 119420 119430 119434 119470 119474 119476 119478 119482 119528 119563 119565 119566 119566 119585 119589 119592 119599 119629 119682 119719 119719 119741 119746 119756 119787 119792 119793 119819 119856 119859 119874 119879 119881 119934 119935 119944 119947 119957 119974 119979 119981 119986 119997 120012 120030 120036 120052 120059 120061 120085 120102 120140 120143 120145 120150 120178 120187 120194 120199 120207 120217 120226 120226 120229 120259 120264 120273 120274 120304 120305 120337 120360 120380 120383 120414 120431 120451 120460 120466 120485 120486 120502 120513 120524 120524 120549 120584 120599 120621 120627 120636 120674 120677 120692 120697 120699 120718 120722 120734 120805 120812 120829 120840 120848 120866 120880 120884 120916 120945 120949 120951 120976 121004 121006 121025 121032 121042 121048 121060 121077 121088 121112 121118 121140 121142 121154 121166 121212 121227 121232 121244 121286 121305 121322 121325 121342 121355 121379 121410 121422 121434 121439 121476 121493 121493 121498 121511 121533 121541 121549 121574 121580 121588 121590 121664 121685 121692 121711 121725 121744 121750 121753 121770 121773 121773 121816 121823 121839 121877 121890 121892 121893 121900 121905 121905 121908 121913 121951 121979 121999 122004 122006 122007 122019 122021 122051 122064 122067 122081 122087 122094 122106 122113 122122 122164 122164 122190 122201 122201 122238 122280 122316 122334 122340 122361 122365 122374 122425 122446 122483 122500 122513 122543 122570 122573 122589 122610 122628 122641 122657 122672 122699 122724 122727 122728 122766 122889 122897 122899 122906 122941 122956 122971 122987 123006 123058 123067 123077 123084 123112 123115 123125 123150 123168 123175 123198 123207 123209 123214 123222 123250 123275 123275 123283 123323 123330 123361 123385 123411 123444 123447 123451 123473 123489 123539 123549 123551 123562 123579 123583 123589 123591 123595 123601 123613 123615 123622 123625 123653 123656 123668 123674 123694 123720 123728 123761 123779 123795 123799 123829 123846 123850 123871 123934 123948 123951 123960 123990 123990 124018 124038 124051 124052 124074 124108 124149 124188 124188 124191 124207 124230 124237 124243 124243 124255 124263 124283 124288 124299 124305 124323 124331 124342 124344 124346 124369 124390 124393 124417 124424 124429 124432 124433 124454 124454 124482 124509 124527 124561 124565 124572 124588 124609 124611 124611 124617 124634 124644 124661 124661 124673 124681 124681 124685 124691 124755 124771 124776 124777 124779 124789 124790 124793 124807 124850 124884 124896 124902 124906 124909 124912 124936 124938 124943 124947 124958 124982 124984 124992 125005 125013 125022 125024 125042 125046 125047 125050 125051 125060 125096 125120 125122 125127 125139 125188 125193 125207 125238 125261 125275 125280 125348 125361 125362 125403 125427 125433 125434 125478 125502 125542 125553 125570 125571 125579 125589 125601 125603 125635 125645 125676 125680 125693 125719 125743 125747 125763 125763 125769 125770 125773 125776 125788 125796 125819 125829 125838 125845 125871 125896 125898 125903 125939 125947 125973 125978 125992 126018 126021 126080 126116 126127 126141 126147 126157 126184 126204 126225 126251 126271 126322 126337 126338 126352 126386 126396 126422 126427 126431 126463 126467 126485 126514 126535 126546 126554 126571 126583 126586 126588 126598 126603 126604 126617 126630 126652 126670 126677 126685 126694 126728 126731 126747 126756 126769 126787 126797 126798 126802 126823 126837 126849 126870 126871 126886 126894 126940 126946 126948 126981 126990 127015 127123 127125 127128 127130 127139 127142 127144 127145 127168 127185 127199 127218 127219 127235 127238 127252 127256 127265 127277 127277 127279 127314 127319 127353 127356 127391 127393 127420 127429 127437 127441 127449 127494 127508 127509 127514 127561 127566 127583 127583 127584 127600 127629 127641 127662 127674 127698 127726 127734 127765 127768 127768 127778 127778 127786 127792 127813 127828 127856 127857 127915 127917 127973 127984 127994 127995 127997 127997 128009 128017 128033 128086 128089 128090 128101 128113 128138 128159 128187 128197 128199 128210 128221 128266 128274 128278 128286 128288 128290 128290 128291 128310 128333 128335 128344 128361 128383 128386 128390 128407 128414 128445 128446 128450 128478 128480 128509 128516 128531 128533 128540 128551 128552 128557 128557 128611 128627 128633 128636 128675 128698 128704 128717 128722 128745 128765 128767 128773 128787 128795 128796 128805 128807 128809 128829 128862 128866 128873 128881 128900 128902 128915 128966 128985 128988 129020 129030 129032 129037 129039 129076 129082 129083 129083 129095 129102 129103 129103 129103 129109 129146 129157 129165 129174 129183 129187 129190 129224 129225 129235 129265 129286 129308 129339 129350 129395 129396 129398 129432 129434 129451 129459 129470 129532 129560 129565 129573 129589 129624 129628 129646 129649 129677 129703 129705 129726 129727 129730 129739 129744 129777 129780 129796 129799 129881 129887 129906 129907 129932 129934 129938 129954 129968 129969 130037 130044 130046 130048 130062 130085 130095 130111 130112 130125 130146 130172 130176 130181 130206 130210 130214 130257 130263 130272 130277 130310 130317 130328 130332 130342 130415 130416 130473 130500 130524 130558 130559 130583 130584 130584 130587 130621 130623 130628 130630 130651 130661 130670 130680 130680 130697 130722 130729 130758 130777 130784 130789 130818 130847 130851 130852 130868 130877 130902 130917 130939 130947 130950 130965 130965 130982 131010 131022 131039 131064 131066 131071 131089 131090 131120 131161 131163 131167 131170 131190 131220 131228 131229 131235 131243 131251 131273 131287 131293 131338 131341 131358 131386 131395 131396 131398 131426 131439 131442 131443 131445 131456 131480 131547 131567 131601 131631 131653 131726 131735 131743 131749 131750 131754 131760 131799 131808 131823 131841 131853 131857 131873 131882 131884 131896 131900 131902 131921 131921 131933 131947 131958 131964 131998 132016 132020 132023 132034 132103 132145 132147 132159 132169 132175 132184 132207 132255 132273 132279 132287 132293 132294 132350 132372 132373 132377 132378 132388 132415 132425 132432 132463 132465 132487 132488 132489 132503 132512 132521 132525 132529 132536 132559 132589 132616 132620 132655 132660 132676 132690 132695 132698 132700 132711 132728 132732 132739 132775 132789 132815 132818 132834 132835 132844 132862 132884 132908 132933 133005 133006 133025 133026 133054 133089 133125 133136 133140 133154 133188 133190 133190 133193 133196 133215 133216 133219 133238 133249 133271 133274 133281 133286 133288 133301 133325 133356 133433 133440 133457 133458 133460 133498 133511 133531 133533 133549 133562 133563 133568 133586 133598 133605 133664 133673 133705 133738 133741 133769 133770 133791 133819 133828 133858 133881 133980 133992 134002 134005 134010 134021 134038 134049 134065 134071 134099 134105 134105 134110 134136 134163 134191 134202 134219 134227 134236 134265 134267 134268 134291 134298 134302 134309 134331 134332 134341 134363 134487 134495 134509 134534 134549 134556 134563 134587 134588 134589 134592 134626 134627 134632 134703 134713 134729 134743 134745 134754 134756 134770 134773 134774 134789 134801 134806 134808 134829 134860 134902 134908 134915 134958 134978 134983 135008 135024 135026 135051 135084 135099 135133 135136 135142 135179 135182 135199 135208 135275 135278 135301 135303 135305 135305 135306 135319 135323 135337 135343 135365 135408 135418 135421 135443 135465 135500 135505 135508 135526 135563 135564 135585 135598 135599 135627 135631 135635 135641 135681 135694 135716 135721 135722 135731 135785 135787 135796 135806 135820 135823 135880 135889 135907 135924 135924 135925 135961 135980 136018 136031 136036 136045 136068 136106 136130 136132 136170 136175 136200 136206 136225 136226 136227 136239 136244 136279 136326 136340 136369 136370 136380 136416 136428 136430 136434 136438 136444 136450 136452 136453 136457 136494 136502 136503 136530 136535 136591 136596 136605 136639 136667 136671 136674 136683 136686 136696 136704 136744 136758 136761 136779 136797 136805 136808 136814 136843 136856 136866 136875 136896 136901 136973 136975 136979 137029 137036 137051 137110 137127 137130 137181 137187 137190 137192 137213 137214 137226 137231 137236 137243 137245 137269 137303 137323 137325 137326 137329 137333 137334 137340 137354 137372 137385 137395 137422 137429 137437 137466 137502 137513 137516 137526 137549 137554 137559 137582 137582 137595 137625 137629 137639 137647 137664 137665 137668 137685 137731 137733 137742 137778 137816 137819 137845 137852 137887 137894 137895 137905 137929 137930 137951 137951 137959 137997 138008 138019 138040 138041 138046 138051 138075 138092 138104 138111 138139 138155 138166 138191 138195 138202 138206 138208 138213 138228 138267 138283 138292 138298 138331 138333 138364 138423 138452 138471 138527 138538 138553 138558 138572 138587 138642 138654 138658 138688 138693 138832 138833 138839 138859 138861 138864 138880 138900 138905 138912 138984 138984 138988 138995 139035 139045 139088 139089 139094 139135 139160 139180 139189 139203 139230 139232 139248 139253 139263 139268 139270 139283 139295 139302 139360 139378 139378 139408 139410 139416 139431 139432 139434 139435 139443 139466 139476 139512 139522 139546 139572 139578 139624 139636 139644 139646 139686 139717 139733 139735 139739 139750 139768 139769 139784 139798 139807 139814 139846 139850 139856 139857 139885 139908 139913 139957 139976 139977 139999 140046 140046 140049 140050 140066 140069 140072 140086 140092 140094 140104 140128 140129 140159 140160 140170 140181 140188 140209 140212 140218 140221 140222 140257 140271 140282 140284 140296 140297 140314 140328 140341 140350 140352 140357 140378 140404 140424 140440 140456 140473 140487 140527 140531 140570 140582 140606 140614 140614 140618 140641 140666 140674 140680 140685 140690 140724 140737 140751 140757 140766 140769 140785 140802 140812 140819 140822 140828 140857 140858 140860 140883 140891 140943 140947 140963 140964 140981 141005 141012 141024 141040 141066 141071 141074 141085 141106 141108 141112 141115 141130 141131 141138 141144 141165 141170 141200 141209 141215 141227 141235 141268 141279 141289 141292 141311 141342 141343 141349 141352 141385 141386 141387 141396 141404 141435 141485 141490 141491 141527 141548 141557 141571 141572 141574 141575 141578 141595 141599 141603 141606 141662 141671 141674 141686 141695 141698 141723 141740 141741 141745 141753 141791 141792 141798 141809 141815 141820 141874 141897 141903 141906 141910 141911 141912 141924 141933 141960 141965 142019 142025 142058 142063 142066 142113 142134 142138 142145 142163 142179 142205 142225 142239 142294 142315 142318 142327 142334 142344 142380 142397 142397 142411 142421 142438 142443 142475 142553 142553 142566 142573 142578 142595 142595 142602 142623 142657 142668 142676 142683 142705 142707 142714 142717 142731 142736 142828 142840 142899 142918 142919 142924 142929 142943 142945 142954 142975 143054 143100 143165 143193 143220 143224 143233 143234 143259 143264 143301 143335 143337 143338 143366 143369 143375 143403 143424 143425 143438 143442 143474 143478 143483 143488 143492 143494 143502 143605 143634 143667 143675 143683 143731 143734 143741 143754 143763 143774 143784 143805 143822 143839 143840 143842 143858 143892 143895 143911 143949 143952 143968 143969 143978 143998 144003 144051 144065 144066 144077 144079 144087 144143 144175 144179 144190 144213 144270 144273 144291 144292 144308 144328 144340 144342 144371 144378 144379 144401 144405 144406 144411 144412 144422 144442 144465 144470 144472 144482 144503 144510 144510 144524 144531 144548 144560 144565 144600 144600 144610 144631 144640 144671 144711 144721 144729 144730 144764 144784 144798 144811 144811 144818 144818 144820 144843 144848 144848 144880 144882 144902 144915 144917 144918 144923 144925 144931 144937 144944 144970 145042 145065 145088 145095 145102 145120 145127 145131 145142 145176 145190 145198 145225 145225 145225 145225 145233 145248 145260 145270 145273 145290 145304 145319 145324 145326 145340 145372 145393 145406 145408 145434 145448 145458 145468 145470 145483 145486 145508 145511 145536 145553 145561 145565 145570 145573 145586 145610 145614 145617 145622 145667 145668 145675 145709 145713 145730 145737 145737 145780 145801 145867 145892 145898 145898 145904 145913 145940 145968 145992 146022 146037 146048 146050 146078 146109 146127 146141 146146 146170 146202 146211 146215 146221 146249 146249 146290 146297 146304 146321 146344 146359 146362 146398 146401 146433 146467 146475 146482 146496 146499 146500 146514 146542 146567 146569 146588 146594 146605 146616 146642 146649 146667 146684 146686 146700 146719 146719 146731 146745 146746 146747 146764 146769 146807 146813 146817 146831 146857 146880 146886 146903 146955 146962 146971 147009 147032 147041 147092 147093 147110 147129 147146 147181 147200 147225 147234 147239 147254 147272 147275 147307 147328 147334 147351 147360 147370 147383 147416 147417 147434 147440 147442 147477 147495 147499 147519 147521 147530 147540 147562 147563 147572 147592 147599 147613 147627 147649 147653 147656 147656 147663 147692 147705 147746 147749 147768 147786 147803 147809 147816 147826 147832 147845 147852 147864 147921 147946 147947 147950 147956 147973 147980 147999 148011 148026 148039 148042 148046 148094 148114 148152 148179 148201 148225 148271 148289 148297 148312 148315 148349 148371 148373 148416 148441 148456 148457 148462 148468 148471 148482 148483 148497 148512 148525 148536 148555 148555 148556 148592 148599 148605 148611 148620 148658 148668 148686 148730 148756 148758 148780 148800 148806 148847 148860 148883 148892 148900 148900 148900 148902 148913 148930 148963 148983 148988 148992 149007 149041 149043 149051 149055 149076 149088 149111 149138 149148 149163 149198 149224 149239 149260 149263 149273 149275 149319 149322 149340 149346 149377 149387 149388 149402 149414 149415 149420 149466 149473 149495 149533 149537 149567 149583 149597 149630 149642 149665 149684 149687 149695 149696 149715 149722 149725 149733 149750 149790 149797 149798 149829 149878 149880 149888 149896 149942 149971 149975 149981 149982 149992 149996 150000"


#from sys import stdin
import timeit
import resource
import sys
start = timeit.default_timer()

#lines = [x for x in stdin.read().split('\n')]

constMaxSlices, constTypesOfPizza = [int(x) for x in line1.split()]
pizzaList = [int(x) for x in line2.split()]

pizzaList.reverse()

checkedIndexList = [] #global variable set
listOfFinalRecurLists = [] #global variable set
maxCurrSumSet = [] #global variable set
globalIndex = 0
boolIsFound = False

print ("No. of pizza :", constTypesOfPizza)
noOfCombi = 2 ** constTypesOfPizza
print ("No. of Combination :", noOfCombi)
print ("Pizza Sets :", pizzaList)


def findMyLocBinarySearch(l, r, intSearchNum): 
    global boolIsFound

    # Check base case
    if boolIsFound == True:
        return True
    elif r >= l: 
  
        midIndex = l + ((r - l) // 2)

        checkedIndexList.append(midIndex)
        
        midNumBinList = convertToBinList(midIndex)
        midNumPizzaList = convertToPizzaList(midNumBinList)

        midNumNextBinList = convertToBinList(midIndex+1)
        midNumNextPizzaList = convertToPizzaList(midNumNextBinList)
  
        midNumDec = sum(midNumPizzaList)
        midNumNextDec = sum(midNumNextPizzaList)

        # If element is present at the middle itself 
        if midNumDec == intSearchNum: 
            listOfFinalRecurLists.append(midIndex)
            boolIsFound = True
          
        # If element is smaller than mid and midNext, then it can only be present in left sublist 
        elif midNumDec > intSearchNum and midNumNextDec > intSearchNum: 
            findMyLocBinarySearch(l, midIndex-1, intSearchNum)
  
        # If element is larger than mid and midNext, then it can only be present in right sublist 
        elif midNumDec < intSearchNum and midNumNextDec < intSearchNum: 
            findMyLocBinarySearch(midIndex+1, r, intSearchNum)
  
        # Else the element can only be present in both sublist 
        else: 
            findMyLocBinarySearch(l, midIndex-1, intSearchNum)
            findMyLocBinarySearch(midIndex+1, r, intSearchNum)
  
    else: 
        # Element is not present in the list
        midNumDec = sum(convertToPizzaList(convertToBinList(l)))
        if midNumDec <= intSearchNum:
            listOfFinalRecurLists.append(l)

        #return l

def convertToBinList(index):
    binStr = "{0:{fill}{width}b}".format(index, fill='0', width=constTypesOfPizza)
    return [int(d) for d in binStr]

def convertToPizzaList(binList):
    tempList = []

    for i, num in enumerate(binList):        
        if num == 1:
            tempList.append(pizzaList[i])

    return tempList

def updateGlobalMaxCurrSumSet(tempIndex):
    global maxCurrSumSet
    global globalIndex

    tempPizzaList = convertToPizzaList(convertToBinList(tempIndex))
    
    intSumTempList = sum(tempPizzaList)
    if sum(maxCurrSumSet) <= intSumTempList <= constMaxSlices:
        maxCurrSumSet = tempPizzaList
        globalIndex = tempIndex


def checkFinalRecurLists():
    for index in listOfFinalRecurLists:
        updateGlobalMaxCurrSumSet(index)


def getFinalListIndexAnswer():
    
    finalpizzaListIndex = []
    tempPizzaList = pizzaList[::-1]
    
    for num1 in reversed(maxCurrSumSet):
        for i, num2 in enumerate(tempPizzaList):
            if num1 == num2:
                finalpizzaListIndex.append(i)
                tempPizzaList[i] = 0
                break
    
    return ' '.join(map(str, [ele for ele in finalpizzaListIndex]))


def printResult(message):
    print (message)
    print ("Sum:", sum(maxCurrSumSet), 'Time:', round(timeit.default_timer() - start, 5), "seconds.", "at position:", globalIndex)
    print ("------------------------Answer------------------------")
    print (len(maxCurrSumSet))
    print (getFinalListIndexAnswer())
    print ("------------------------------------------------------")




# Start Function call
defaultSysRecursionLimit = sys.getrecursionlimit()
sys.setrecursionlimit(constTypesOfPizza+100) # set to constTypesOfPizza

findMyLocBinarySearch(0, noOfCombi-1, constMaxSlices)

sys.setrecursionlimit(defaultSysRecursionLimit) # set to system default

print ("no. of checked indexes in binary search:", len(checkedIndexList))
print ("Final Recur list length:", len(listOfFinalRecurLists))

checkFinalRecurLists()
printResult ("========================================binary search result...========================================")

print ("Memory usage:", resource.getrusage(resource.RUSAGE_SELF).ru_maxrss)
```
