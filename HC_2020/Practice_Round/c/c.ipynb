```
line1 = "4500 50" #@param {type:"string"}
line2 = "7 12 12 13 14 28 29 29 30 32 32 34 41 45 46 56 61 61 62 63 65 68 76 77 77 92 93 94 97 103 113 114 114 120 135 145 145 149 156 157 160 169 172 179 184 185 189 194 195 195" #@param {type:"string"}


#from sys import stdin
import timeit
import resource
import math
import sys
start = timeit.default_timer()

#lines = [x for x in stdin.read().split('\n')]

constMaxSlices, constTypesOfPizza = [int(x) for x in line1.split()]
pizzaList = [int(x) for x in line2.split()]

pizzaList.reverse()

maxCurrSumSet = [] #global variable set
globalIndex = 0
checkedIndexList = [] #global variable set

print ("No. of pizza :", constTypesOfPizza)
noOfCombi = 2 ** constTypesOfPizza
print ("No. of Combination :", noOfCombi)


def findMyLocBinarySearch(l, r, intSearchNum): 
  
    # Check base case 
    if r >= l: 
  
        midIndex = l + ((r - l) // 2)

        checkedIndexList.append(midIndex)

        binStr = "{0:{fill}{width}b}".format(midIndex, fill='0', width=constTypesOfPizza)
        binStrList = [int(d) for d in binStr]
        
        midNumDec = convertToPizzaList(binStrList, midIndex)
  
        # If element is present at the middle itself 
        if midNumDec == intSearchNum: 
            return midIndex 
          
        # If element is smaller than mid, then it can only be present in left sublist 
        elif midNumDec > intSearchNum: 
            return findMyLocBinarySearch(l, midIndex-1, intSearchNum) 
  
        # Else the element can only be present in right sublist 
        else: 
            return findMyLocBinarySearch(midIndex+1, r, intSearchNum) 
  
    else: 
        # Element is not present in the list 
        return -1

        
def findMyLocLinearSearch(myMidIndex):

    intRangeLimit = (int((math.log2(constTypesOfPizza) ** 2) * constTypesOfPizza))
    intLowerRange = 0 if ((myMidIndex-intRangeLimit) < 0) else (myMidIndex-intRangeLimit)
    intUpperRange = noOfCombi if ((myMidIndex+intRangeLimit) > noOfCombi) else (myMidIndex+intRangeLimit)

    print ("--------------------------------------")
    print ("Linear Search Limit:", intLowerRange, "to", intUpperRange)
    print ("--------------------------------------")

    for i in range(intLowerRange, intUpperRange):
      
        if sum(maxCurrSumSet) == constMaxSlices:
            break

        binStr = "{0:{fill}{width}b}".format(i, fill='0', width=constTypesOfPizza)
        binStrList = [int(d) for d in binStr]
        
        midNumDec = convertToPizzaList(binStrList, i)


def convertToPizzaList(binList, myIndex):
    global maxCurrSumSet
    global globalIndex

    tempList = []

    for i, num in enumerate(binList):        
        if num == 1:
            tempList.append(pizzaList[i])
    
    intSumTempList = sum(tempList)
    if (intSumTempList > sum(maxCurrSumSet)) and (intSumTempList <= constMaxSlices):
        maxCurrSumSet = tempList
        globalIndex = myIndex

        printTempResult ("-------------intermediate search result..-------------")

    #print (binList, tempList, intSumTempList)

    return intSumTempList


def getFinalListIndex(pizzaList, finalpizzaList):
    
    finalpizzaListIndex = []
    tempPizzaList = pizzaList[::-1]
    
    for num1 in reversed(finalpizzaList):
        for i, num2 in enumerate(tempPizzaList):
            if num1 == num2:
                finalpizzaListIndex.append(i)
                tempPizzaList[i] = 0
                break
    
    return ' '.join(map(str, [ele for ele in finalpizzaListIndex]))

def printTempResult(message):
    print (message)
    print ("Sum:", sum(maxCurrSumSet), 'Time:', round(timeit.default_timer() - start, 5), "seconds.")

def printResult(message):
    print (message)
    print ("Sum:", sum(maxCurrSumSet), "at position:", globalIndex, 'Time:', round(timeit.default_timer() - start, 5), "seconds.")
    print ("------------------------Answer------------------------")
    print (len(maxCurrSumSet))
    print (getFinalListIndex(pizzaList, maxCurrSumSet))
    print ("------------------------------------------------------")




# Start Function call
print ("========================================binary search started..========================================")
sys.setrecursionlimit(constTypesOfPizza+100) # set to constTypesOfPizza

mySearchLoc = findMyLocBinarySearch(0, noOfCombi-1, constMaxSlices)
print (mySearchLoc)

sys.setrecursionlimit(1000) # set to system default

printResult ("========================================binary search result...========================================")
print ("no. of checked indexes in binary search:", len(checkedIndexList))

# If optimal set not found in Binary search: do linear searching around globalMidIndex
if mySearchLoc == -1:
    print ("========================================linear search started..========================================")
    findMyLocLinearSearch(globalIndex)
# Else optimal set found in Binary search: maxCurrSumSet must have alrady updated in findMyLocBinarySearch
else:
    print ("rare case")

printResult ("========================================linear search result...========================================")

print (resource.getrusage(resource.RUSAGE_SELF).ru_maxrss)
```
