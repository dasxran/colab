```
line1 = "4500 50" #@param {type:"string"}
line2 = "7 12 12 13 14 28 29 29 30 32 32 34 41 45 46 56 61 61 62 63 65 68 76 77 77 92 93 94 97 103 113 114 114 120 135 145 145 149 156 157 160 169 172 179 184 185 189 194 195 195" #@param {type:"string"}


#from sys import stdin
import timeit
import resource
import sys
start = timeit.default_timer()

#lines = [x for x in stdin.read().split('\n')]

constMaxSlices, constTypesOfPizza = [int(x) for x in line1.split()]
pizzaList = [int(x) for x in line2.split()]

pizzaList.reverse()

checkedIndexList = [] #global variable set
listOfFinalRecurLists = [] #global variable set
maxCurrSumSet = [] #global variable set
globalIndex = 0
boolIsFound = False

print ("No. of pizza :", constTypesOfPizza)
noOfCombi = 2 ** constTypesOfPizza
print ("No. of Combination :", noOfCombi)
print ("Pizza Sets :", pizzaList)


def findMyLocBinarySearch(l, r, intSearchNum): 
    global boolIsFound

    # Check base case
    if boolIsFound == True:
        return True
    elif r >= l: 
  
        midIndex = l + ((r - l) // 2)

        checkedIndexList.append(midIndex)
        
        midNumBinList = convertToBinList(midIndex)
        midNumPizzaList = convertToPizzaList(midNumBinList)

        midNumNextBinList = convertToBinList(midIndex+1)
        midNumNextPizzaList = convertToPizzaList(midNumNextBinList)
  
        midNumDec = sum(midNumPizzaList)
        midNumNextDec = sum(midNumNextPizzaList)

        # If element is present at the middle itself 
        if midNumDec == intSearchNum: 
            listOfFinalRecurLists.append(midIndex)
            boolIsFound = True
          
        # If element is smaller than mid and midNext, then it can only be present in left sublist 
        elif midNumDec > intSearchNum and midNumNextDec > intSearchNum: 
            findMyLocBinarySearch(l, midIndex-1, intSearchNum)
  
        # If element is larger than mid and midNext, then it can only be present in right sublist 
        elif midNumDec < intSearchNum and midNumNextDec < intSearchNum: 
            findMyLocBinarySearch(midIndex+1, r, intSearchNum)
  
        # Else the element can only be present in both sublist 
        else: 
            findMyLocBinarySearch(l, midIndex-1, intSearchNum)
            findMyLocBinarySearch(midIndex+1, r, intSearchNum)
  
    else: 
        # Element is not present in the list
        midNumDec = sum(convertToPizzaList(convertToBinList(l)))
        if midNumDec <= intSearchNum:
            listOfFinalRecurLists.append(l)

        #return l

def convertToBinList(index):
    binStr = "{0:{fill}{width}b}".format(index, fill='0', width=constTypesOfPizza)
    return [int(d) for d in binStr]

def convertToPizzaList(binList):
    tempList = []

    for i, num in enumerate(binList):        
        if num == 1:
            tempList.append(pizzaList[i])

    return tempList

def updateGlobalMaxCurrSumSet(tempIndex):
    global maxCurrSumSet
    global globalIndex

    tempPizzaList = convertToPizzaList(convertToBinList(tempIndex))
    
    intSumTempList = sum(tempPizzaList)
    if sum(maxCurrSumSet) <= intSumTempList <= constMaxSlices:
        maxCurrSumSet = tempPizzaList
        globalIndex = tempIndex


def checkFinalRecurLists():
    for index in listOfFinalRecurLists:
        updateGlobalMaxCurrSumSet(index)


def getFinalListIndexAnswer():
    
    finalpizzaListIndex = []
    tempPizzaList = pizzaList[::-1]
    
    for num1 in reversed(maxCurrSumSet):
        for i, num2 in enumerate(tempPizzaList):
            if num1 == num2:
                finalpizzaListIndex.append(i)
                tempPizzaList[i] = 0
                break
    
    return ' '.join(map(str, [ele for ele in finalpizzaListIndex]))


def printResult(message):
    print (message)
    print ("Sum:", sum(maxCurrSumSet), 'Time:', round(timeit.default_timer() - start, 5), "seconds.", "at position:", globalIndex)
    print ("------------------------Answer------------------------")
    print (len(maxCurrSumSet))
    print (getFinalListIndexAnswer())
    print ("------------------------------------------------------")




# Start Function call
print ("========================================binary search started..========================================")
defaultSysRecursionLimit = sys.getrecursionlimit()
sys.setrecursionlimit(constTypesOfPizza+100) # set to constTypesOfPizza

findMyLocBinarySearch(0, noOfCombi-1, constMaxSlices)

sys.setrecursionlimit(defaultSysRecursionLimit) # set to system default

print ("no. of checked indexes in binary search:", len(checkedIndexList))
print ("Final Recur list length:", len(listOfFinalRecurLists))

checkFinalRecurLists()
printResult ("========================================binary search result...========================================")

print ("Memory usage:", resource.getrusage(resource.RUSAGE_SELF).ru_maxrss)
```
